## External Dataset 2 - Tripti

```python
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/Data Science with Python/Group project 2/soi_data.csv')   # reading the csv data into dataframe
df.head()
```

```python
df.shape                                                                          # checking the number of rows and columns in df dataframe
```

```python
df['zipcode'].nunique()                                                            # checking the number of unique zipcodes
```

Number of rows are more than unique zipcodes which means that data is not at zip code level

```python
df.info()                                                                          # checking the data types of columns
```

```python
df.head(-10)                                                                      # inspecting the last 10 rows of df dataframe
```

```python
df1= df[~((df['zipcode'] == 0) | (df['zipcode'] ==99999))]                        # removing the garbage data with zipcode 0 and 99999
df1.shape
```

```python
# aggergating the data at zip code level 
# picked 10 columns which intutively made sense and added value. 
#aggregating for both sum and mean as mean is needed for A02650 and rest all need sum

df2=df1.groupby(['zipcode'])[['N1','mars1','MARS2','MARS4','ELF','CPREP','PREP','DIR_DEP','N2','A02650']].agg(['sum','mean']).reset_index()

df2.shape
```

```python
df2.head()
```

```python
# only selecting the required columns

df3= df2.iloc[:,[0,1,3,5,7,9,11,15,17,20]]

df3.head()
```

```python
# renaming the columns

df3.rename(columns = {'N1' : 'returns', 'mars1': 'single_returns','MARS2':'joint_returns','MARS4':'head_household_returns','ELF':'elec_returns','CPREP':'comp_prep_returns','N2':'individuals',
                      'A02650':'avg_income_amount'}, inplace=True)

df3.head()
```

```python
# checking the datatype and non-null values for all columns

df3.info()
```

```python
# exporting the clean data to a csv file to finally import in the main file

df3.to_csv(r'/content/drive/MyDrive/Data Science with Python/Group project 2/soi_data_clean.csv', index=False)
```
